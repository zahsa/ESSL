```python
concatted_dfs = []
for batch_size in merged["batch size"].unique():
    bs = merged[merged["batch size"] == batch_size]
    # acuuracy
    c10 = [[col, bs[bs["data set"] == "cifar10"][col].max()] for col in bs[bs["data set"] == "cifar10"].columns if 'acc' in col]
    svhn = [[col, bs[bs["data set"] == "svhn"][col].max()] for col in bs[bs["data set"] == "svhn"].columns if 'acc' in col]
    c10_cols = ["dataset"] + [c[0] for c in c10]
    c10_vals = [["cifar10"] + [c[1] for c in c10]]
    c10_df = pd.DataFrame(c10_vals, columns=c10_cols)[["dataset", "test acc (supervised baseline)",
                                                         "test acc (ssl baseline)", "test acc (so)",
                                                         "test acc (mo)",'test acc (ssl optimized)']]
    c10_df.columns = [ 

                        "Dataset",
                        "Supervised",  
                        "SSL (default)", 
                        "SSL (SO)", 
                        "SSL (MO)", 
                        "SSL (optimized)"
    ]
    svhn_cols = ["dataset"] + [c[0] for c in c10]
    svhn_vals = [["svhn"] + [c[1] for c in svhn]]
    svhn_df = pd.DataFrame(svhn_vals, columns=svhn_cols)[["dataset", "test acc (supervised baseline)",
                                                         "test acc (ssl baseline)", "test acc (so)",
                                                         "test acc (mo)",'test acc (ssl optimized)']]
    svhn_df.columns = [ 

                        "Dataset",
                        "Supervised",  
                        "SSL (default)", 
                        "SSL (SO)", 
                        "SSL (MO)", 
                        "SSL (optimized)"
    ]
    print("Accuracy", f"bs={batch_size}")
    c_df = pd.concat([c10_df, svhn_df])
    display(c_df)
    concatted_dfs.append([batch_size, "acc", c_df])
    
    # loss
    c10 = [[col, bs[bs["data set"] == "cifar10"][col].max()] for col in bs[bs["data set"] == "cifar10"].columns if 'loss' in col]
    svhn = [[col, bs[bs["data set"] == "svhn"][col].max()] for col in bs[bs["data set"] == "svhn"].columns if 'loss' in col]
    c10_cols = ["dataset"] + [c[0] for c in c10]
    c10_vals = [["cifar10"] + [c[1] for c in c10]]
    c10_df = pd.DataFrame(c10_vals, columns=c10_cols)[["dataset", "test loss (supervised baseline)",
                                                         "test loss (ssl baseline)", "test loss (so)",
                                                         "test loss (mo)",'test loss (ssl optimized)']]
    c10_df.columns = [ 

                        "Dataset",
                        "Supervised",  
                        "SSL (default)", 
                        "SSL (SO)", 
                        "SSL (MO)", 
                        "SSL (optimized)"
    ]
    svhn_cols = ["dataset"] + [c[0] for c in c10]
    svhn_vals = [["svhn"] + [c[1] for c in svhn]]
    svhn_df = pd.DataFrame(svhn_vals, columns=svhn_cols)[["dataset", "test loss (supervised baseline)",
                                                         "test loss (ssl baseline)", "test loss (so)",
                                                         "test loss (mo)",'test loss (ssl optimized)']]
    svhn_df.columns =[ 

                        "Dataset",
                        "Supervised",  
                        "SSL (default)", 
                        "SSL (SO)", 
                        "SSL (MO)", 
                        "SSL (optimized)"
    ]
    print("Loss", f"bs={batch_size}")
    c_df = pd.concat([c10_df, svhn_df])
    display(c_df)
    concatted_dfs.append([batch_size, "loss", c_df])
```

```python
for bs, metric, df in concatted_dfs:
    print(df.round(2).to_latex(caption=f"best found {metric}, bs={bs}", label=f"best_{metric}_{bs}"))
```

```python
algos = merged['algo'].unique()
tables = []
for dset in merged['data set'].unique():
    for bs in merged['batch size'].unique():
        print(dset, bs, "acc")
        ds_bs_acc = merged[(merged['data set'] == dset) & (merged['batch size'] == bs)][[col for col in merged.columns if 'acc' in col or col == 'algo']]
        ds_bs_acc.columns =  ["algo",
                        "SSL (SO)", 
                        "SSL (MO)",  
                        "SSL (default)", 
                        "Supervised",
                        "SSL (optimized)"]
        try:
            ds_bs_acc["SSL (optimized)"] = ds_bs_acc["SSL (optimized)"].astype("float64")
        except:
            pass
        ds_bs_acc = ds_bs_acc.set_index('algo')
        ds_bc_acc = ds_bs_acc.round(2).transpose()
        display(ds_bs_acc.round(2).transpose())
        tables.append([dset, bs, "acc", ds_bc_acc.round(2).transpose()])
        
        print(dset, bs, "loss")
        ds_bs_loss = merged[(merged['data set'] == dset) & (merged['batch size'] == bs)][[col for col in merged.columns if 'loss' in col or col == 'algo']]
        ds_bs_loss.columns =  ["algo",
                        "SSL (SO)", 
                        "SSL (MO)",  
                        "SSL (default)", 
                        "Supervised",
                        "SSL (optimized)"]
        try:
            ds_bs_loss["SSL (optimized)"] = ds_bs_loss["SSL (optimized)"].astype("float64")
        except:
            pass
        ds_bs_loss = ds_bs_loss.set_index('algo')
        ds_bc_loss = ds_bs_loss.round(2).transpose()
        display(ds_bs_loss.round(2).transpose())
        tables.append([dset, bs, "loss", ds_bc_loss.round(2).transpose()])
        
print(algos)
```

```python
for dset, bs, metric, df in  tables:
    print(df.transpose().to_latex(caption=f"{dset} {metric}, bs={bs}", label=f"full_{metric}_{bs}"))
```
